```python
# Change column names for clarity -- Log GDP per capita to just GDP per capita

df.rename(columns = {'Log GDP per capita' : 'GDP per capita'}, inplace = True)
```

```python
# Check for rename

df.head(1)
```

```python
# Check data types -- change year to object

df['year'] = df['year'].astype('object')
```

```python
# checking data type change

df.info()
```

```python
# Check for mixed type data

for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# no mixed type data to address
# no columns need to be dropped, all relevant to analysis
```
