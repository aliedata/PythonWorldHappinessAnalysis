```python
# adding formatting and styling to heat map

f = plt.figure(figsize=(11, 11))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.shape[1]), df.columns, fontsize=9, rotation=45)
plt.yticks(range(df.shape[1]), df.columns, fontsize=12)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('World Happiness Correlation Matrix', fontsize=16)
```

# 6. Create a categorical plot and interpret the results.

```python
# creating a histogram with GDP

sns.histplot(df['Healthy life expectancy at birth'], bins = 10, kde = True)
```

```python
df.loc[df['Healthy life expectancy at birth'] < 55, 'LE category'] = 'Low Life Expectancy'
```

```python
df.loc[(df['Healthy life expectancy at birth'] >= 55) & (df['Healthy life expectancy at birth'] < 70), 'LE category'] = 'Average Life Expectancy'
```

```python
df.loc[df['Healthy life expectancy at birth'] >= 70, 'LE category'] = 'High Life Expectancy'
```

```python
df['LE category'].value_counts(dropna = False)
```

```python
# creating a categorical plot in seaborn using the categories created above / TAKING TOO LONG, MADE COMPRABLE SCATTERPLOT BELOW

sns.set(style='ticks')
g = sns.catplot(x='Healthy life expectancy at birth', y='year', hue='LE category', data=df)
```

```python
sns.scatterplot(data=df, x='Healthy life expectancy at birth', y='year', hue='LE category')
```

Discussion: 

The chart above shows that most life expecancy (LE) is in the middle range for all countries year over year (2006-2020). Happily, low life expectancies per country have appeared to decrease over time, while high life expectancy has increased. Very surprising to see this shift in only 14 years!
